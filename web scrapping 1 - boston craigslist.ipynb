```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
url='https://boston.craigslist.org/search/npo'
our_data={}
job_no=0
while True:   
    
    response=requests.get(url)
    data=response.text
    soup=BeautifulSoup(data,'html.parser')
    jobs=soup.find_all('p',attrs={'class':'result-info'})

    for job in jobs:
        title=job.find('a',attrs={'class':'result-title'}).text
        adress_tag=job.find('span',attrs={'class':'result-hood'})
        adress=adress_tag.text if adress_tag else 'N/A'
        date=job.find('time',attrs={'class':'result-date'}).text
        link=job.find('a',attrs={'class':'result-title'}).get('href')
         
        job_response=requests.get(link)
        job_data=job_response.text
        job_soup=BeautifulSoup(job_data,'html.parser')
        job_description=job_soup.find('section',{'id':'postingbody'}).text
        
        #compensation and employment type 
        job_tag=job_soup.find('p',{'class':'attrgroup'})
        job_attr=job_tag.text if job_tag else 'N/A'
        job_no+=1
        #collect all data inside the dict,which the key is job_no OR title
        our_data[job_no]=[title,adress,date,link,job_attr,job_description]
        
        
        #print(job_no)
        print('Title:',title ,'\nlocation:',adress,'\ndate:',date ,'\nlink:',link,'\nJob Description:',job_attr,job_description,'\n________')
        
    #to get the url for next page and replay all the previos steps on it
    url_tag = soup.find('a',{'title':'next page'})
    if url_tag.get('href'):
        url= 'https://boston.craigslist.org' + url_tag.get('href')
        print(url)
    else:
        break
        
print('Total Jobs:',job_no)
the_jobs=pd.DataFrame.from_dict(our_data , orient='index',columns=['title','location','Date','link','job attributes','job description'])

```

```python
#save to CSv file
the_jobs.to_csv('The_Jobss.csv')
```
